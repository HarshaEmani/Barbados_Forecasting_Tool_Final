```python
from DB_Utils import fetch_data, fetch_forecast_data_with_options
import pandas as pd
import plotly.express as px

start_date = '2024-06-01'
end_date = '2024-06-30'

baseload_preds = fetch_forecast_data_with_options(1, start_date, end_date, columns=['actual_value', 'forecast_value'], model='LSTM_Baseload', scenario='24hr')
change_load_preds = fetch_forecast_data_with_options(1, start_date, end_date, columns=['forecast_value'], model='LSTM_Change_in_Load', scenario='24hr')

change_load_preds = change_load_preds.rename(columns={'forecast_value': 'change_load_forecast_value'})
baseload_preds = baseload_preds.rename(columns={'actual_value': 'baseload_actual_value', 'forecast_value': 'baseload_forecast_value'})

combined_preds = pd.concat([baseload_preds[['baseload_actual_value', 'baseload_forecast_value']], change_load_preds['change_load_forecast_value']], axis=1)
combined_preds = combined_preds.dropna()

px.line(combined_preds)
```

```python
actual_value_np = combined_preds['baseload_actual_value'].values.reshape(-1, 24)

change_in_load = actual_value_np[1:] - actual_value_np[:-1]
back_to_base = actual_value_np[:-1] + change_in_load

px.line(change_in_load.flatten()).show()
px.line(back_to_base.flatten()).show()

change_in_load
```
